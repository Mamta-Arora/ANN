Keras is by default wrapped in tensorflow version 2.0 or above. if you are using tensorflow 2.0 or above version then there is no need to import keras seprately.

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
```

Sample DataSet Prepartion

```python
from sklearn import datasets
```

```python
x,y = datasets.make_moons(n_samples = 1000)
print (x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

## Create Model

```python
model =Sequential()
model.add(Dense(10,input_shape=(2,),activation = 'relu'))
model.add(Dense(10,activation = 'relu'))
model.add(Dense(1,activation ='sigmoid'))
```

## Fit the Data

```python
sgd = SGD(lr =0.01)
model.compile(optimizer = 'sgd', metrics = 'accuracy', loss = 'binary_crossentropy')
model.fit(x,y,epochs =50, verbose = True)
```

## Plot the Decision Boundary

```python
plot_decision_regions(x,y,model)
```
